```python
# funcion que extrae la categoria del venue, otra forma
def get_category_type(row):
    try:
        categories_list = row['categories']    
    except:
        categories_list = row['venue.categories']    
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name'] 
```

```python
df_f['categories'] = df_f.apply(get_category_type, axis=1) #get the category name 
df_f
```
